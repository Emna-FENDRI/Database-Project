### Building small standalone tables:

1) Safety equipment :

```python
#### create the table containing all the safety equipments

p_safety_equipment = parties[["party_safety_equipment_1", "party_safety_equipment_2"]].drop_duplicates()
p_safety_equipment.rename(columns={"party_safety_equipment_1":"safety_equipment_1", "party_safety_equipment_2":"safety_equipment_2"}, inplace=True)

v_safety_equipment = victims[["victim_safety_equipment_1", "victim_safety_equipment_2"]].drop_duplicates()
v_safety_equipment.rename(columns={"victim_safety_equipment_1":"safety_equipment_1", "victim_safety_equipment_2":"safety_equipment_2"}, inplace=True)

safety_equipment = pd.concat([p_safety_equipment, v_safety_equipment]).drop_duplicates()
safety_equipment.insert(loc=0, column="seid", value=np.arange(len(safety_equipment.index)))


safety_equipment
```

2) Associate factor:

```python
### create the table containing all the combination of associate factor

associate_factor = parties[["other_associate_factor_1", "other_associate_factor_2"]].dropna().drop_duplicates()
associate_factor.insert(loc=0, column="afid", value=np.arange(len(associate_factor.index)))
associate_factor
```

3) Weather:

```python
### create the table containing all the weathers

weather = collisions.weather_1.combine_first(collisions.weather_2).dropna().unique()
weather = pd.DataFrame.from_dict({'afid' : range(len(weather)), 'weather' : np.sort(weather), 'code_letter':['A', 'B', 'E', 'F', 'C', 'D', 'G']})
weather
```

### Building bigger standalone tables:

1) Vehicle:

2) Road state:

```python
def build_table(df, cols, file_name):
    """
    build a table containing the combination of different values of cols 
    in df and write it in file_name
    """
    
    print("############# " + file_name + " ###################")
    file = df[cols].drop_duplicates(ignore_index=True)
    file.insert(loc=0, column=file_name+"_id", value=file.index)
    display(file)
    
    ### ratio of full line : 
    print(f"> {round(len(file.dropna(how='any').index) * 100 / len(file.index),  2)} % of the dataset are described by the 3 components")


    ##### What to do with all those NaN?
    for col in cols:
        print(f"> proportion of NaN for {col} :") 
        print(round(100 - (len(file[col].dropna(how='any').index) * 100 / len(file.index)),  2), "%")
    
    print("> Reduction of entity number : ")
    print(f"original length : {len(df[cols].index)}, without duplicates : {len(file.index)} ")
    return file
    
```

```python
## create a road_state

road_state = build_table(collisions, ['road_condition_1', 'road_condition_2', 'road_surface', 'lighting'], 'road_state')
```

```python
# create the table containing all the vehicle

vehicle = build_table(parties, ['statewide_vehicle_type', 'vehicle_make', 'vehicle_year'], "vehicle")
```

```python
## create the road state

road_state = build_table(collisions, ['road_condition_1', 'road_condition_2', 'road_surface', 'lighting'], 'road_state')
```

```python
## create the location

location = build_table(collisions, ['county_city_location', 'jurisdiction', 'population', 'location_type'], 'location')
```

```python
### create the pcf

pcf = build_table(collisions, ['pcf_violation', 'pcf_violation_category', 'pcf_violation_subsection', 'primary_collision_factor'], 'pcf')
```

### Creation of the relations:

1) has_weather:

```python
w1 = collisions[['case_id', 'weather_1']].dropna(how='any')
w1.rename(columns={'weather_1' : 'weather'}, inplace=True)
w2 = collisions[['case_id', 'weather_2']].dropna(how='any')
w2.rename(columns={'weather_2' : 'weather'}, inplace=True)
```

```python
display(w1)
display(w2)
display(weather)
```

```python
#### create the table relating the case_id to weather_1 and weather_2

has_weather_1 = w1.merge(weather, on=['weather'])[['case_id', 'afid']].drop_duplicates()
has_weather_1 = has_weather_1.rename(columns={'afid' : 'weather_id_1'}, inplace=False)
has_weather_2 = w2.merge(weather, on=['weather'])[['case_id', 'afid']].drop_duplicates()
has_weather_2 = has_weather_2.rename(columns={'afid' : 'weather_id_2'}, inplace=False)
```

```python
display(has_weather_1)
```

2) has_factor :

```python
### create the table relating the party_id to the associate factor

has_factors = parties[['id', 'other_associate_factor_1', 'other_associate_factor_2']]
has_factors = associate_factor.merge(has_factors, on=['other_associate_factor_1', 'other_associate_factor_2'])[['id', 'afid']].drop_duplicates()
```

```python
### create the table containing the table relating the parties to the safety equipment

party_used_equipment = parties[['id', 'party_safety_equipment_1', 'party_safety_equipment_2']]
party_used_equipment.rename(columns={'party_safety_equipment_1':'safety_equipment_1', 'party_safety_equipment_2':'safety_equipment_2'}, inplace=True)
party_used_equipment = party_used_equipment.merge(safety_equipment, on=['safety_equipment_1', 'safety_equipment_2'], how='left')[['id', 'seid']].drop_duplicates()
party_used_equipment
```

4) victim_has_equipment:

```python
### same but for the victim

victim_used_equipment = victims[['id', 'victim_safety_equipment_1', 'victim_safety_equipment_2']]
victim_used_equipment.rename(columns={'victim_safety_equipment_1':'safety_equipment_1', 'victim_safety_equipment_2':'safety_equipment_2'}, inplace=True)
victim_used_equipment = victim_used_equipment.merge(safety_equipment, on=['safety_equipment_1', 'safety_equipment_2'], how='left')[['id', 'seid']].drop_duplicates()
victim_used_equipment
```

5) has_road_state:

```python
## relate the road_state to the collisions

has_road_state = collisions[['case_id', 'road_condition_1', 'road_condition_2', 'road_surface', 'lighting']]
has_road_state = road_state.merge(has_road_state, on=['road_condition_1', 'road_condition_2', 'road_surface', 'lighting']).drop_duplicates()
has_road_state = has_road_state[['road_state_id', 'case_id']].drop_duplicates()
display(has_road_state)
```

```python
## relate the collisions to a pcf

has_pcf = collisions[['case_id', 'pcf_violation', 'primary_collision_factor', 'pcf_violation_category', 'pcf_violation_subsection']]
has_pcf = pcf.merge(has_pcf, on=['pcf_violation', 'primary_collision_factor', 'pcf_violation_category', 'pcf_violation_subsection']).drop_duplicates()
has_pcf = has_pcf[['pcf_id', 'case_id']].drop_duplicates()
display(has_pcf)
```
